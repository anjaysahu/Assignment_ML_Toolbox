Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

#### Answer -

**1. scatterplot -** When we want to plot the combination point of any data that time we use scatterplot.    
__2. displot -__ When we want to plot the frequency of any data that time we use displot.   
**3. jointplot -** When we want to plot "scatterplot" and "displot" both data at once that time we use jointplot.    
**4. relplot -** When we want to plot any data in a group that time we use relplot.      
**5. catplot -** When we want to plot categorical and numerical data that time we use catplot.         

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x ="timepoint" and y = "signal" for different events and regions.   
Note: timepoint, signal, event, and region are columns in the fmri dataset.

#### Solution -

```python
import seaborn as sns
```

```python
fmri = sns.load_dataset("fmri")
```

```python
fmri
```

```python
sns.lineplot(x ="timepoint", y = "signal",data=fmri,hue="event")
```

```python
sns.lineplot(x ="timepoint", y = "signal",data=fmri,hue="region")
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x ='pclass', y = 'age' and y = 'fare'.       
Note: pclass, age, and fare are columns in the titanic dataset. 

#### Solution -

```python
titanic = sns.load_dataset("titanic")
```

```python
titanic
```

```python
sns.boxplot(x ="pclass", y = "age",data=titanic)
```

```python
sns.boxplot(x ="pclass", y = "fare",data=titanic)
```
